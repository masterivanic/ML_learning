12) Résoudre le système d’équations 3x0 + x1 = 9 et x0 + 2x1 = 8

```python
# Coefficients des équations
coefficients = np.array([[3, 1], [1, 2]])

# Termes constants
constants = np.array([9, 8])

# Résolution du système d'équations
solution = np.linalg.solve(coefficients, constants)

print("Solution du système d'équations :", solution)
```

38) (Défi) On considère le tableau suivant qui représente la liste des ages des élèves d'une classe de terminal:

age = [19,17,21,18,19,19, 22,23,20,18,21,24, 19,22,20,18,22,21, 19,20,19,23,21,18, 20,19,20,20,19,20]

Ecrire un programme en python basé sur le module python statistics, qui permet de calculer les différents paramètres statistique du tableau: moyenne, variance, ecart-type, mode, mediane...

```python
import statistics as st

age = [19,17,21,18,19,19,
 22,23,20,18,21,24,
19,22,20,18,22,21,
19,20,19,23,21,18,
20,19,20,20,19,20]

moyenne = st.mean(age)
print(f"Moyenne = {moyenne}")

# clacul de la variance et de l'ecart-type
#-------------------------------------------
var , ecart = st.variance(age) , st.stdev(age)
print(f"Variance = {var}")
print(f"Ecart-type = {ecart}")

#-------------------------------------------
# determination de la médiane statistique
#-------------------------------------------
mediane = st.median(age)
print(f"mediane = {mediane}")

# determination du mode statistique
#-------------------------------------------
mode = st.mode(age)
print(f"Mode = {mode}")
```

40) Si l'on considère le vecteur [1, 2, 3, 4, 5], comment construire un nouveau vecteur avec 3 zéros consécutifs intercalés entre chaque valeur ?



```python
Z = np.array([1,2,3,4,5])
nz = 3
Z0 = np.zeros(len(Z) + (len(Z)-1)*(nz))
Z0[::nz+1] = Z
print(Z0)
```
